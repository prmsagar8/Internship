### Q2: In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

#### 1st Page data

```python
urls = []

for i in driver.find_elements_by_xpath('//a[@class="a-link-normal s-no-outline"]'):
    urls.append(i.get_attribute('href'))
```

```python
len(urls)
```

```python
urls # Product url variable
```

```python
#page 1 variables

Brand_Name1 = []
NameOfTheProduct_1= []
Rating_1= []
No_ofRatings_1= []
Price1 = []
Return_Exchange1 = []
ExpectedDelivery1 = []
Availability_1= []
OtherDetails1 = []
```

```python
for i in urls:
    driver.get(i)
    time.sleep(3)
    
    #fetching brand name data
    try:
        names1 = driver.find_element_by_xpath('//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]')
        Brand_Name1.append(names1.text.replace('\n', 'new line'))
    except:
        Brand_Name1.append('-')
        
    #fetching Name Of The Product
    try:
        nameproduct1 = driver.find_element_by_xpath('//*[@id="productTitle"]')
        NameOfTheProduct_1.append(nameproduct1.text.replace('\n', 'new line'))
    except:
        NameOfTheProduct_1.append('-')
        
    # fetching rating
    try:
        Rating1 = driver.find_element_by_xpath('//*[@id="a-popover-content-7"]/div/div/div/div[1]/span')
        Rating_1.append(Rating1.text.replace('\n', 'new line'))
    except:
        Rating_1.append('-')
        
    # fetching No of Rating
    try:
        NoofRatings1 = driver.find_element_by_xpath('//*[@id="acrCustomerReviewText"]')
        No_ofRatings_1.append(NoofRatings1.text.replace('\n', 'new line'))
    except:
        No_ofRatings_1.append('-')
        
    # fetching Price
    try:
        Prices1 = driver.find_element_by_xpath('//*[@id="priceblock_ourprice"]')
        Price1.append(Prices1.text.replace('\n', 'new line'))
    except:
        Price1.append('-')
        
    # fetching Return_Exchange1
    try:
        Exchange1 = driver.find_element_by_xpath('//*[@id="RETURNS_POLICY"]/span/div[2]/a')
        Return_Exchange1.append(Exchange1.text.replace('\n','new line'))
    except:
        Return_Exchange1.append('-')
        
    # fetching Expected Delivery
    try:
        ExpectedDeliverys1 = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[12]/div/div/b')
        ExpectedDelivery1.append(ExpectedDeliverys1.replace('\n','new line'))
    except:
        ExpectedDelivery1.append('-')
        
    #fetching Availability_1
    try: 
        Availabilitys_1 = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[20]/div/div/div/div[2]/div[1]/div/div/table/tbody/tr[4]/td')
        Availability_1.append(Availabilitys_1.replace('\n','new line'))
    except:
        Availability_1.append('-')
    
    #fetching OtherDetails1
    try:
        OtherDetail1 = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[42]/div/ul')
        OtherDetails1.append(OtherDetail1.replace('\n','new line'))
    except:
        OtherDetails1.append('-')
```

### Q3: Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
import urllib
from selenium import webdriver

driver.get('https://images.google.com')
```

### scrapping fruits images

```python
# google page

#search field
google_inputField = driver.find_element_by_xpath('//*[@class="gLFyf gsfi"]')
```

```python
#input scraping fruits
google_inputField.send_keys('fruits')
```

```python
#enter
google_inputField.send_keys(Keys.RETURN)
```

```python
#Scroll browser page to bottom
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
time.sleep(10)
```

```python
#creating a directory to save images
folder_name = 'fruits'
if not os.path.isdir(folder_name):
    os.makedirs(folder_name)

def download_image(url, folder_name, num):

    # write image to file
    reponse = requests.get(url)
    if reponse.status_code==200:
        with open(os.path.join(folder_name, str(num)+".jpg"), 'wb') as file:
            file.write(reponse.content)
```

```python
page_html = driver.page_source
pageSoup = bs4.BeautifulSoup(page_html, 'html.parser')
containers = pageSoup.findAll('div',{'class':"bRMDJf islir"})
```

```python
len_containers = len(containers)
print('Found %s image containers'%(len_containers))
```

```python
for i in range(1, len_containers+1):
    if i% 25==0:
        continue
        

    xPath = """//*[@id="islrg"]/div[1]/div[%s]"""%(i)
    
    previewImageXPath = """//*[@id="islrg"]/div[1]/div[%s]/a[1]/div[1]/img"""%(i)
    previewImageElement = driver.find_element_by_xpath(previewImageXPath)
    previewImageURL = previewImageElement.get_attribute("src")

    driver.find_element_by_xpath(xPath).click()
    
    timeStarted = time.time()
    while True:

        imageElement = driver.find_element_by_xpath("""//*[@id="Sva75c"]/div/div/div[3]/div[2]/c-wiz/div/div[1]/div[1]/div[2]/div[1]/a/img""")
        imageURL= imageElement.get_attribute('src')
        
        if imageURL != previewImageURL:
            #print("actual URL", imageURL)
            break

        else:
            #making a timeout if the full res image can't be loaded
            currentTime = time.time()

            if currentTime - timeStarted > 10:
                print("Timeout! Will download a lower resolution image and move onto the next one")
                break
                

#Downloading image
    try:
        download_image(imageURL, folder_name, i)
        print("Downloaded element %s out of %s total. URL: %s" % (i, len_containers + 1, imageURL))
        
    except:
        print("Couldn't download an image %s, continuing downloading the next one"%(i))
```

### Q8  Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
#url
driver.get('http://www.forbes.com/')
```

```python
#click to billionaires Url
billioners_list = 'https://www.forbes.com/billionaires/'
driver.get(billioners_list)
```

```python
#variables
Rank8 = []
name8 = []
net_worth8 = []
age8 = []
citizenship8 = []
source8 = []
industry8 = []
```

```python
for i in billioners_list:

    try:
        url_8= driver.find_element_by_xpath('//div[@class="next-page"]')#clicks start
        
        
        #fetching rank
        Rank = driver.find_elements_by_xpath('//div[@class="rank"]')

        #using loop to collect rank data
        for i in Rank:
            l = i.text
            Rank8.append(l)
            
        #fetching Name
        name = driver.find_elements_by_xpath('//div[@class="personName"]')

        #using loop to collect name data
        for i in name:
            l = i.text
            name8.append(l)
        
        #fetching net worth
        networth = driver.find_elements_by_xpath('//div[@class="netWorth"]')

        #using loop to collect net worth data
        for i in networth:
            l = i.text
            net_worth8.append(l)    
            
        #fetching Age
        age_8 = driver.find_elements_by_xpath('//div[@class="age"]')

        #using loop to collect agee data
        for i in age_8:
            l = i.text
            age8.append(l)   
            
        #fetching citizenship
        citizen = driver.find_elements_by_xpath('//div[@class="countryOfCitizenship"]')

        #using loop to collect citizenship data
        for i in citizen:
            l = i.text
            citizenship8.append(l)    
            
        #fetching source
        source = driver.find_elements_by_xpath('//div[@class="source-column"]')

        #using loop to collect source data
        for i in source:
            l = i.text
            source8.append(l)    
            
        #fetching industry
        indust = driver.find_elements_by_xpath('//div[@class="category"]')

        #using loop to collect industry data
        for i in indust:
            l = i.text
            industry8.append(l)    
            
            
        url_8.click() #click on ---> Next page... button    
        time.sleep(3)
        
        
    except: 
        pass
        break
```

```python
#checking for data length
len(Rank8), len(name8), len(net_worth8), len(age8), len(citizenship8), len(source8), len(industry8)
```

```python
#creating data frame
df_billioners = pd.DataFrame()
df_billioners['Rank']=Rank8
df_billioners['Name']=name8
df_billioners['Net Worth']=net_worth8
df_billioners['Age']=age8
df_billioners['Citizenship']=citizenship8
df_billioners['Source']=source8
df_billioners['Industry']=industry8
df_billioners
```
