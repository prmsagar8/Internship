### Q6. Scrape the details of top 100 songs on billboard.com.

Url = https://www.billboard.com/

You have to find the following details:
* A) Song name
* B) Artist name
* C) Last week rank
* D) Peak rank
* E) Weeks on board

Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
#exception to run entire scrape data using xpath clicks, nested exception and for_loops.
try:
    #url
    driver.get('https://www.billboard.com/')
    
    
    #chart page
    chartpage= driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/header/div/ul/li[1]')
    chartpage.click()

    # THE HOT 100 page
    hot100 = driver.find_element_by_xpath('//*[@id="main"]/div[2]/div/div[1]/a/div[2]/div[2]/div[1]')
    hot100.click()
    

    #variables
    Song_name = []
    Artist_name = []
    Last_week_rank = []
    Peak_rank = []
    Weeks_on_board = []


    #exceptions for scrape data
    try:
        #Song Name _ loop & xpath
        for i in driver.find_elements_by_xpath('//span[@class="chart-element__information__song text--truncate color--primary"]'):
            Song_name.append(i.text)
    
        #artist Name_ xpath & loop
        for i in driver.find_elements_by_xpath('//span[@class="chart-element__information__artist text--truncate color--secondary"]'):
            Artist_name.append(i.text)
        
        #last week rank _ xpath & loop
        for i in driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--last"]'):
            Last_week_rank.append(i.text)
    
        #Peak_rank _ xpath & loop
        for i in driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--peak"]'):
            Peak_rank.append(i.text)
        
        #last week rank _ xpath & loop
        for i in driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--week"]'):
            Weeks_on_board.append(i.text)        
    
    except NoSuchElementException:
        pass
except NoSuchElementException:
    pass
```

```python
#creating data frame
df_100_hot = pd.DataFrame()
df_100_hot['Song Name'] = Song_name
df_100_hot['Artist Name'] = Artist_name
df_100_hot['Last Week Rank'] = Last_week_rank
df_100_hot['Peak Rank'] = Peak_rank
df_100_hot['Week on Board'] = Weeks_on_board
df_100_hot
```

### Q8. Scrape the details of Highest selling novels.

Url = https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/

You have to find the following details:
* A) Book name
* B) Author name
* C) Volumes sold
* D) Publisher
* E) Genre

```python
#url
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/')
```

```python
#variables
book_name = []
author_name = []
volumes = []
publisher = []
genre = []

try:
    #fetching names
    for i in driver.find_elements_by_xpath('//td[2][@class="left"]'):
        x= i.text.replace('tr', 'td')
        book_name.append(x)
        
    #fetching author names
    for i in driver.find_elements_by_xpath('//td[3][@class="left"]'):
        x= i.text.replace('tr', 'td')
        author_name.append(x)
        
    #fetching volumes
    for i in driver.find_elements_by_xpath('//td[4][@class="left"]'):
        x= i.text.replace('tr', 'td')
        volumes.append(x)
        
    #fetching publisher
    for i in driver.find_elements_by_xpath('//td[5][@class="left"]'):
        x= i.text.replace('tr', 'td')
        publisher.append(x)
        
    #fetching genre
    for i in driver.find_elements_by_xpath('//td[@class="last left"]'):
        x= i.text.replace('tr', 'td')
        genre.append(x)
        
except NoSuchElementExceptions:
    pass
```

```python
#checking length of the data
len(book_name),len(author_name),len(volumes),len(publisher),len(genre)
```

```python
#creating data frames
df_TOP100Books = pd.DataFrame()
df_TOP100Books['Title'] = book_name
df_TOP100Books['Author'] = author_name
df_TOP100Books['Volume Sales'] = volumes
df_TOP100Books['Publisher'] = publisher
df_TOP100Books['Genre'] = genre
df_TOP100Books
```

### Q9. Scrape the details most watched tv series of all time from imdb.com.

You have to find the following details:

* B) Year span
* C) Genre
* D) Run time
* E) Ratings
* F) Votes

```python
try:
    #url
    driver.get('https://www.imdb.com/list/ls095964455/')

    #variables
    titlename = []
    year_span = []
    Genere = []
    Run_time = []
    Ratings = []
    votes = []

    try:
        #titles
        for i in driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a'):
            titlename.append(i.text)
        
        #year_span
        for i in driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/span[2]'):
            year_span.append(i.text)
        
        #Genere
        for i in driver.find_elements_by_xpath('//span[@class="genre"]'):
            Genere.append(i.text)
        
        #Run_time
        for i in driver.find_elements_by_xpath('//span[@class="runtime"]'):
            Run_time.append(i.text)
        
        #Ratings
        for i in driver.find_elements_by_xpath('//*[@id="main"]/div[*]/div[3]/div[3]/div[*]/div[2]/div[1]/div[1]/span[2]'):
            Ratings.append(i.text)
        
        #votes
        for i in driver.find_elements_by_xpath('//*[@id="main"]/div/div[3]/div[3]/div[*]/div[2]/p[4]/span[2]'):
            votes.append(i.text)
        
        
    except NoSuchElementExceptions:
        pass
    
except:
    pass
```

```python
try:
    #url
    driver.get('https://archive.ics.uci.edu/')
    #time.sleep(3)
    
    #click on --->  view all data sets
    click_10 = driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]/a')
    click_10.click()
    

    #variable
    Name = []
    Data_type = []
    Task = []
    Attribute_type = []
    No_of_instances = []
    No_of_attribute = []
    Year = []

    try:
        #fetching names
        for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[*]/td[1]'):
            Name.append(i.text)
        
        #fetching Data_type
        for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[*]/td[2]'):
            Data_type.append(i.text)
        
        #fetching Task
        for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[*]/td[3]'):
            Task.append(i.text)
        
        #fetching Attribute_type
        for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[*]/td[4]'):
            Attribute_type.append(i.text)
        
        #fetching No_of_instances
        for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[*]/td[5]'):
            No_of_instances.append(i.text)
        
        #fetching No_of_attribute
        for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[*]/td[6]'):
            No_of_attribute.append(i.text)
        
        #fetching Year
        for i in driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[*]/td[7]'):
            Year.append(i.text)        
        
    except:
        pass
    
except:
    pass
```
